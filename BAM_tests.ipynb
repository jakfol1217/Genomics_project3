```python
for pilecolumn in bamfile.pileup("chr1", 0, 10400):
    print(pilecolumn.nsegments)
    print(pilecolumn.reference_pos)
    print(pilecolumn.get_mapping_qualities())
    #print(pilecolumn.reference_id)
    #print(pilecolumn.reference_name)
    print(pilecolumn.get_num_aligned())
    print(pilecolumn.get_query_names())
    print(pilecolumn.get_query_positions())
    print(str(pilecolumn))
    print("----------------")
    
```

```python
for pileupcolumn in bamfile.pileup("chr1", 0, 11000):
    print("\ncoverage at base %s = %s" % (pileupcolumn.pos, pileupcolumn.n))
    for pileupread in pileupcolumn.pileups:
        if not pileupread.is_del and not pileupread.is_refskip:
            # query position is None if is_del or is_refskip is set.
            print('\tbase in read %s = %s' %
                  (pileupread.alignment.query_name,
                   pileupread.alignment.query_sequence[pileupread.query_position]))
```
